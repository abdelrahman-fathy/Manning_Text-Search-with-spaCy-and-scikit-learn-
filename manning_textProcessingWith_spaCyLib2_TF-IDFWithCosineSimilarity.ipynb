```python
def build_Tf(vocabs, document_string):
    Tf = {}
    for word in document_string:
        if word in vocabs.keys():
            #print(f'word:{word}, Counter(document_string):{Counter(document_string)[word]}, vocabs[word]: {vocabs[word]}, ration:{Counter(document_string)[word] /vocabs[word]} ')
            Tf[word] = Counter(document_string)[word] /vocabs[word]
    return Tf
#print(data[0]['tokenized_text'])
build_Tf(token_counts, data[1]['tokenized_text'])
```
